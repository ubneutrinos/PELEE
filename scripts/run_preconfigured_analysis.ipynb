```python
from microfit.fileio import from_json
import matplotlib.pyplot as plt

# read results from file
two_hypo_results = from_json(os.path.join("../full_ana_output", "two_hypo_result_new_constraints_smooth_detvar.json"))
minimum = np.min(two_hypo_results["samples_h0"])
minimum = min(minimum, np.min(two_hypo_results["samples_h1"]))
maximum = np.max(two_hypo_results["samples_h0"])
maximum = max(maximum, np.max(two_hypo_results["samples_h1"]))

n_trials = len(two_hypo_results["samples_h0"])
bin_edges = np.linspace(minimum, maximum, int(np.sqrt(n_trials)))


fig, ax = plt.subplots(figsize=(5, 4), constrained_layout=True)
ax.hist(two_hypo_results["samples_h0"], bins=bin_edges, histtype="step", density=True, label="H0")
ax.hist(two_hypo_results["samples_h1"], bins=bin_edges, histtype="step", density=True, label="H1")
ax.axvline(x=two_hypo_results["ts_median_h1"], color="k", linestyle="--", label=f"Median H1\np-val: {two_hypo_results['median_pval']*100:0.3f}%")
ax.legend()
ax.set_xlabel(r"$\Delta \chi^2$")
ax.set_ylabel("Probability density")
ax.set_title("Sensitivity, Runs 1-5, Reco. Shower Energy")
fig.savefig(os.path.join("../full_ana_output", "two_hypo_result_new_constraints_smooth_detvar.pdf"))
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=0)
x = rng.normal(0, 1, 400)
h, bin_edges = np.histogram(x, bins=np.linspace(-1, 3, 20), density=True)

print(f"Sum of h: {np.sum(h)}")

filter = np.array([0.1, 0.3, 1.0, 0.3, 0.1])
filter /= np.sum(filter)

def filter_with_reflection(arr, filter):
    """Filter the array with the given filter, reflecting the array at the edges."""

    # pad the array with zeros
    arr_padded = np.pad(arr, (len(filter) // 2, len(filter) // 2), mode="reflect")

    # apply the filter
    arr_filtered = np.convolve(arr_padded, filter, mode="valid")

    return arr_filtered

h_filtered = filter_with_reflection(h, filter)
print(f"Sum of h_filtered: {np.sum(h_filtered)}")

def repeat_last(arr):
    return np.concatenate([arr, [arr[-1]]])

plt.step(
    bin_edges,
    repeat_last(h),
    where="post",
    label="Data",
    color="black",
)

plt.step(
    bin_edges,
    repeat_last(h_filtered),
    where="post",
    label="Filtered",
    color="red",
)
```
