```python
import sys, os 
sys.path.append("../../")
import numpy as np
```

```python
from microfit.analysis import MultibandAnalysis
import logging

config_file = "../../config_files/full_ana_with_detvars.toml"

analysis = MultibandAnalysis.from_toml(config_file, logging_level=logging.DEBUG, overwrite_cached_df=False)
```

```python
analysis.plot_sidebands(
    include_multisim_errors=True,
    add_precomputed_detsys=True,
    separate_figures=True,
    show_data_mc_ratio=True,
    show_chi_square=True
)
```

```python
plot_dir = "full_ana_output/plots/"
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
analysis.plot_correlation(ax=ax, add_precomputed_detsys=False)
ax.set_title("Correlation matrix w/o detector systematics")
fig.savefig(os.path.join(plot_dir, "correlation_matrix_no_detsys.pdf"))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
analysis.plot_correlation(ax=ax, add_precomputed_detsys=True)
ax.set_title("Correlation matrix w/ detector systematics")
fig.savefig(os.path.join(plot_dir, "correlation_matrix_with_detsys.pdf"))
```

```python
analysis.plot_correlation(ms_columns=[], include_unisim_errors=False, add_precomputed_detsys=True, as_correlation=True, as_fractional=False)
```

```python
from microfit.parameters import Parameter, ParameterSet

two_hypo_result = analysis.two_hypothesis_test(
    h0_params=ParameterSet([Parameter("signal_strength", 0.0)]),
    h1_params=ParameterSet([Parameter("signal_strength", 1.0)]),
    sensitivity_only=True,
    add_precomputed_detsys=True,
    n_trials=100000,
)
```

```python
plt.figure(figsize=(4.5, 4))
bin_edges = np.linspace(-40, 50, 100)
plt.hist(two_hypo_result["samples_h0"], bins=bin_edges, label="H0", histtype="step")
plt.hist(two_hypo_result["samples_h1"], bins=bin_edges, label="H1", histtype="step")
plt.axvline(two_hypo_result["ts_median_h1"], color="black", label=f"Median H1\np-val={two_hypo_result['median_pval']*100:.3f}%")
plt.legend()
plt.semilogy()
plt.xlabel(r"$\Delta \chi^2$")
plt.ylabel("Samples")
plt.title("Runs 1-5 Sensitivity")
plt.savefig(os.path.join(plot_dir, "sensitivity_runs1-5.pdf"))
```
